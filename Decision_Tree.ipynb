```python
#importing the libraries
from sklearn.datasets import load_breast_cancer #loading the breast cancer dataset
from sklearn import tree #loading the tree module for Decision Tree Classifier
from sklearn.model_selection import train_test_split #loading train and test data set splitter 
```

```python
#loading the cancer data set
#this dataset is one of the buit with the sklearn library
cancer = load_breast_cancer()
```

<h1>Cancer dataset</h1>
<p>  This is the dataset provided by (Dr. William H. Wolberg, General Surgery Department, University of Wisconsin) <br>
This data set provides the various features of the a cell which are liste below.<br>
So using these features we see if we can predict if a cell is Malignant(cancer cell/ harmful), Benign(non cancer cell/harmless)
</p>

```python
print("Features:")
print(cancer.feature_names)
print("Sample Data")
print(cancer.data[0])
print("Target Names:")
print(cancer.target_names)
print("Target for the sample:")
print(cancer.target[0])
```

#A look in to the data

```python
print ("Detailed Look")
for val in range(cancer.data[0].size):
  print("%d.%s = %r" % (val,cancer.feature_names[0], cancer.data[0][val]))
print("Sample target data")
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size = .15)
```

```python
dec_tree_clf = tree.DecisionTreeClassifier(random_state = 0)
dec_tree_clf.fit(X_train, y_train)
```

```python
#predicting result in a the test set

index = 6

print("The prediction for:")
for val in range(X_test[index].size):
  print("%s = %r" % (cancer.feature_names[val], X_test[0][val]))

prediction = dec_tree_clf.predict([X_test[0]]);
print("___________________________________")
print("The classification is :" + cancer.target_names[prediction][0])
print("The actual should be :" + cancer.target_names[y_test[index]])
```

```python
print('Accuracy on the training subset: {:, 3f}', format(dec_tree_clf.score(X_train, y_train)))
print('Accuracy on the testing subset: {:, 3f}', format(dec_tree_clf.score(X_test, y_test)))
```
